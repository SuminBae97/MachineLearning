```python
zeros_cnt = df.isnull().sum().sort_values(ascending=False)
percent_zeros = (df.isnull().sum() / df.isnull().count()).sort_values(ascending=False)

missing_data = pd.concat([zeros_cnt, percent_zeros], axis=1, keys=['Total', 'Percent'])
missing_data
```

```python
zero_count = df.isnull().sum().sort_values(ascending=False)
```

```python
df.isnull().sum().sort_values(ascending=False)
zero_patio = (df.isnull().sum() / df.isnull().count()).sort_values(ascending=False)


missing_data = pd.concat([zero_count , zero_patio],axis=1 , keys=['Total',"%pct"])
```
